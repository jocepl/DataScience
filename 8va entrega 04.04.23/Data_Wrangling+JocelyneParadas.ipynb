#### Descripcin de las variables

* **id**: Identificador nico de cada registro


* **name**: Nombre de la cancin


* **popularity**: Popularidad


* **duration_ms**: Duracion en microsegundos


* **explicit**: indica si la cancin contiene lenguaje explcito (0 = no, 1 = s) 


* **artists**: Artista/artistas


* **id_artists**: Identificador de los artistas


* **release_date**: fecha de lanzamiento


* **danceability**: Spotify define esta cualidad, que podra ser traducida como bailiabilidad, midiendo cunto de apta para bailar es la cancin, basndose en una combinacin de elementos musicales que incluye el tempo, la estabilidad rtmica, la intensidad del pulso y la regularidad general (de 0 a 1).


* **key**:	 tonalidad de la cancin (0 = Do, 1 = Do#, 2 = Re, etc.)


* **loudness**:	nivel de sonoridad de la cancin en decibelios (dB)


* **mode**: indica si la cancin est en modo mayor (1) o menor (0)


* **speechiness**: Por speechiness se entiende la presencia de palabras habladas en el tema. Para pistas totalmente habladas, como un audiolibro, un recital de un poema, etc., el valor de este parmetro se acercar a 1.0.


* **instrumentalness**:	Este parmetro da informacin sobre la presencia de elementos vocales en la cancin. Cuanto ms prximo est este valor a 1.0, mayor probabilidad de que el tema no tenga voces.


* **acousticness**: medida de la presencia de elementos acsticos en la cancin (de 0 a 1)


* **tempo**: Velocidad con que se interpreta una composicin musical.


* **valence**: es una medida de la positividad o negatividad de una cancin. Se define como la medida de la "valencia musical" de una pista, que oscila entre 0 y 1, donde 0 es la pista ms triste y 1 es la pista ms feliz o alegre.

### 10 canciones ms populares del 2021

```python
Populares = df.query('popularity > 90', inplace = False).sort_values('popularity', ascending = False)

primeros10 = Populares[0:10]
display(primeros10[['name','popularity', 'artists','release_date']])
```

```python
corr_df = primeros10.drop(["duration_ms",'year'],axis=1).corr(method="pearson")
plt.figure(figsize= (14,6))
heatmap= sns.heatmap(corr_df,annot= True,fmt= ".1g", vmin=-1,vmax=1, center=0,cmap="viridis",linewidths=1,linecolor='black')
heatmap.set_title("Correlacion entre las variables de las canciones mas populares del 2021")
heatmap.set_xticklabels(heatmap.get_xticklabels(),rotation = 90)
```

```python
plt.figure(figsize=(6,2))
sns.regplot(data=primeros10, x="loudness", y="energy", color="c").set(title="Correlacin entre Loudness vs Energy")
```

```python
plt.figure(figsize=(6,2))
sns.regplot(data=primeros10, y="energy", x="acousticness", color="c").set(title="Correlacin entre Acousticness vs Energy")
```

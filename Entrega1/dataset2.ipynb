## EcoBicis en Buenos Aires
***
Informacion de viajes realizados en 2022. 
Estacion de origen y destino, hora de partida y llegada.

En este data set podemos encontrar los siguentes atributos:

* **Id_recorrido:** Clave identificadora del recorrido
* **duracion_recorrido:** Diferencia entre la fecha de origen y la fecha fin en segundos
* **fecha_origen_recorrido:** Fecha de inicio
* **id_estacion_origen:** Id Estacion de salida
* **nombre_estacion_origen:** Nombre estacion de origen
* **direccion_estacion_origen:** direcion escrita de la estacion de origen
* **long_estacion_origen:**	Longitud de la estacion de origen
* **lat_estacion_origen:**	Latitud de la direccion de origen
* **fecha_destino_recorrido:**	Fecha final del recorrido
* **id_estacion_destino:**	Id estacion final
* **nombre_estacion_destino:**	Nombre de la estacion final
* **direccion_estacion_destino:** direcion escrita de la estacion final
* **long_estacion_destino:** Longitud de la estacion final
* **lat_estacion_destino:**	Latitud de la direccion final
* **id_usuario:** id usuario
* **modelo_bicicleta:**	modelo de bicicleta
* **Género:** Género del usuario

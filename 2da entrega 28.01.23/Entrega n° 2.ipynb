# EcoBicis en Buenos Aires
***
Informacion de viajes realizados en 2022. 
Estacion de origen y destino, hora de partida y llegada.

En este data set podemos encontrar los siguentes atributos:

* **Id_recorrido:** Clave identificadora del recorrido
* **duracion_recorrido:** Diferencia entre la fecha de origen y la fecha fin en segundos
* **fecha_origen_recorrido:** Fecha de inicio
* **id_estacion_origen:** Id Estacion de salida
* **nombre_estacion_origen:** Nombre estacion de origen
* **direccion_estacion_origen:** direcion escrita de la estacion de origen
* **long_estacion_origen:**	Longitud de la estacion de origen
* **lat_estacion_origen:**	Latitud de la direccion de origen
* **fecha_destino_recorrido:**	Fecha final del recorrido
* **id_estacion_destino:**	Id estacion final
* **nombre_estacion_destino:**	Nombre de la estacion final
* **direccion_estacion_destino:** direcion escrita de la estacion final
* **long_estacion_destino:** Longitud de la estacion final
* **lat_estacion_destino:**	Latitud de la direccion final
* **id_usuario:** id usuario
* **modelo_bicicleta:**	modelo de bicicleta
* **Género:** Género del usuario

### Uso de las EcoBicis en la cidudad de Buenos Aires por sexo

Podemos ver que las EcoBicis de la ciudad son más utilizadas por el sexo masculino, que por las mujeres y otras personas que no se identifican con los dos géneros o sexos antriormente mencionados.

Así como tambien podemos ver un aumento en el uso de las bicicletas tipo Iconic, por lo que podemos concluir que no hay una preferencia para un tipo de bici en especifico segun el sexo, sino que podemos inferir que otras variables intervienen en la eleccion del tipo de bicicleta, por ejemplo, la cantidad de las bicis tipo Fit en la ciudad.

Por otra parte, podemos ver que tanto el sexo femenino como los que definen su género como "otro", tienen un aumeto en el tiempo de uso de las bicicletas de la ciudad, teniendo un promedio levemente mayor el sexo femenino.

```python
df['Género'] = df['Género'].replace(['MALE'], 'Masculino')
df['Género'] = df['Género'].replace(['FEMALE'], 'Femenino')
df['Género'] = df['Género'].replace(['OTHER'], 'Otro')

generos = df.Género.value_counts()

plt.bar(x=generos.index, height=generos.values, color='orange', edgecolor='grey')
```

```python
sns.countplot(data=df, x='Género', hue="modelo_bicicleta", color='darkorange')
```

```python
sns.catplot(data=df, x="duracion_recorrido_min", y="Género", kind="box")
```

### Cantidad de Bicicletas segun su tipo

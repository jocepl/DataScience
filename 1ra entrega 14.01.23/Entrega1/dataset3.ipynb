## Spotify
***
Este dataset nos brinda información sobre qué canciones son las más populares, cuántas canciones ganaron popularidad en un puntaje de entre 90 y 100, cúal es la duracion de una cancion promedio entre otros datos.

En este data set podemos encontrar los siguentes atributos:

* **id**: Identificador único de cada registro
* **name**: Nombre de la canción
* **popularity**: Popularidad
* **duration_ms**: Duracion en microsegundos
* **explicit**: ¿La cancion es explicita? 0 = No, 1 = Si 
* **artists**: Artista/artistas
* **id_artists**: Identificador de los artistas
* **release_date**: fecha de lanzamiento
* **danceability**: Spotify define esta cualidad, que podría ser traducida como bailiabilidad, midiendo cuánto de apta para bailar es la canción, basándose en una combinación de elementos musicales que incluye el tempo, la estabilidad rítmica, la intensidad del pulso y la regularidad general.
* **key**:	clave del registo
* **loudness**:	nivel de sonido
* **mode**:	modo de la cancion
* **speechiness**: Por speechiness se entiende la presencia de palabras habladas en el tema. Para pistas totalmente habladas, como un audiolibro, un recital de un poema, etc., el valor de este parámetro se acercará a 1.0.
* **instrumentalness**:	Este parámetro da información sobre la presencia de elementos vocales en la canción. Cuanto más próximo esté este valor a 1.0, mayor probabilidad de que el tema no tenga voces.	
* **tempo**: Velocidad con que se interpreta una composición musical.

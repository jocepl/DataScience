## Puntos de Wi-Fi Públicos en Buenos Aires
***
Información y ubicación geográfica de los puntos de red de conexión a internet inalámbrica gratuita ubicados en plazas, parques, sedes comunales y bibliotecas de la Ciudad.

En este data set podemos encontrar los siguentes atributos:

* **Longitud** (number):	Longitud de localización
* **Latitud** (number):	Latitud de localización
* **id** (string):	Identificador único de cada registro
* **nombre** (string):	Nombre de cada elemento de la tabla
* **tipo** (string):	Tipo de elemento
* **etapa**	(string): Etapa de inaguración de los puntos Wi-Fi
* **etapa_obse** (string): 	Nombre de la etapa mas observaciones secundarias
* **estado** (string: Referencia respecto a la disponibilización del servicio
* **subcategoria** (string): Subcategoría del sitio
* **calle_nombre** (string): Nombre de la calle principal
* **calle_altura** (integer)	Altura de calle
* **calle_cruce** (string): Nombre de la calle que intercepta la calle principal
* **barrio** (string): Barrio de localización de la obra
* **comuna** (string): Comuna de localización de la obra
* **codigo_postal** (integer): Código postal
* **codigo_postal_argentino** (integer): Código postal del país
* **objeto** (string): Tipo de objeto
